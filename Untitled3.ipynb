```python
def to_bin(data):
    if isinstance(data, str):
        return ''.join(format(ord(c), '08b') for c in data)
    elif isinstance(data, (bytes, bytearray, np.ndarray)):
        return ''.join(format(b, '08b') for b in data)
    elif isinstance(data, int):
        return format(data, '08b')
    else:
        raise TypeError("Unsupported type.")
```

```python
def encode_lsb(img, secret):
    data = to_bin(secret) + '1111111111111110'
    arr = np.array(img)
    flat = arr.flatten()
    if len(data) > len(flat):
        raise ValueError("Message too big!")
    mask = np.uint8(254)  # 11111110, clears LSB without overflow
    for i, bit in enumerate(data):
        flat[i] = (flat[i] & mask) | np.uint8(int(bit))
    return Image.fromarray(flat.reshape(arr.shape).astype('uint8'))


```

```python
secret = input("Enter secret message: ")
stego = encode_lsb(cover, secret)
stego.save("stego.png")
files.download("stego.png")
```

```python
def decode_lsb(img):
    flat = np.array(img).flatten()
    bits = [str(byte & 1) for byte in flat]
    chars = []
    for i in range(0, len(bits), 8):
        byte = ''.join(bits[i:i+8])
        if byte == '11111111':  # part of terminator
            tail = ''.join(bits[i+8:i+24])
            if tail.endswith('111110'):
                break
        chars.append(chr(int(byte, 2)))
    return ''.join(chars)

# Upload & extract
uploaded = files.upload()
stego_path = list(uploaded.keys())[0]
stego_img = Image.open(stego_path).convert('RGB')
print("Hidden message:", decode_lsb(stego_img))
```
